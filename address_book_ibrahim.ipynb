```python
contacts = {}
trial = 0

while True:  
    print('[1]: Add Contact\n[2]: Edit Contact\n',
        '[3]: Delete Contact\n[4]: Search Contact\n[#]: Exit', sep='')
    
    option = input('Enter option: ')
    
    if option == '1':
        name = input('Enter name: ')
        if name in contacts:
            print(f'{name} already exist')
        else:
            add = input('Enter address: ')
            phones = []
            emails = []

            while True:
                phone = input(f'Enter phone no {len(phones)+1}: ')
                phones.append(phone)

                if input('Do you wish to add more no?(Y/N): ').upper() != 'Y':
                    break

            while True:
                email = input(f'Enter email {len(emails)+1}: ')
                emails.append(email)

                if input('Do you wish to add more no?(Y/N): ').upper() != 'Y':
                    break
                
            contacts[name] = {'add': add, 'phones': phones, 'emails': emails}
        
    elif option == '2':
        name = input('Enter name: ')
        if name in contacts:
            print(f'Name: {name}')
            print(f"Address: {contacts[name]['add']}")
            print(f"Phone Nos: {contacts[name]['phones']}")
            print(f"Emails: {contacts[name]['emails']}")
            
            while True:
                print('[1]: Edit Address\n[2]: Edit Phone Nos\n[3]: Edit Emails\n[#]: Previous Menu')
                option = input('Enter option: ')
                
                if option == '1':
                    old_add = contacts[name]['add']
                    add = input('Enter New Address: ')
                    contacts[name]['add'] = add
                    print(f'{old_add} successfully updated to {add}')
                elif option == '2':
                    while True:
                        print('[1]: Add New\n[2]: Edit\n[3]: Delete\n[#]: Previous Menu')
                        option = input('Enter option: ')
                        
                        if option == '1':
                            new_phone = input('Enter New Phone No: ')
                            contacts[name]['phones'].append(new_phone)
                            print(f'{new_phone} successfully added.')
                        elif option == '2':
                            for i in range(len(contacts[name]['phones'])):
                                print(f"{i}: {contacts[name]['phones'][i]}")
                            
                            index = int(input('Enter option: '))
                            new_phone = input('Enter New Phone No: ')
                            contacts[name]['phones'][index] = new_phone
                            print(f'{new_phone} successfully updated.') 
                            
                        elif option == '3':
                            for i in range(len(contacts[name]['phones'])):
                                print(f"{i}: {contacts[name]['phones'][i]}")
                                
                            index = int(input('Enter option: '))
                            temp = contacts[name]['phones'][index]
                            del contacts[name]['phones'][index]
                            print(f'{temp} successfully deleted.') 
                            
                        elif option == '#':
                            break
                
                elif option == '3':
                    pass
                elif option == '#':
                    break
            
            
        else:
            print(f'{name} does not exist!')
        
    elif option == '3':
        name = input('Enter name: ')
        if name in contacts:
            del contacts[name]
            print(f'{name} successfully deleted!')
        else:
            print(f'{name} does not exist!')
        
    elif option == '4':
        name = input('Enter name: ')
        if name in contacts:
            print(f'Name: {name}')
            print(f"Address: {contacts[name]['add']}")
            print(f"Phone Nos: {contacts[name]['phones']}")
            print(f"Emails: {contacts[name]['emails']}")
        else:
            print(f'{name} does not exist!')
            
    elif option == '#':
        print('Goodbye!!!')
        break
    else:
        print('Invalid option')
        trial += 1
        
        if trial == 3:
            print('Technical Issue')
            break
```
